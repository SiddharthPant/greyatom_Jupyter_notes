```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic/titanic.csv', index_col=0)
df
```

```python
proportions = df['Sex'].value_counts()
# print proportions
proportions.name = ''
proportions.plot.pie(
    labels=['Males', 'Females'],
    explode=(0,0.15),
    startangle=90,
    autopct='%1.1f%%',
#     colors=['b','r'],
    fontsize='14'
)
plt.axis('equal')
plt.tight_layout()
# plt.title('Sex Proportion')
plt.figtext(.5,.97,'Sex Proportion',fontsize=15,ha='center')
# plt.legend()
plt.show()
```
