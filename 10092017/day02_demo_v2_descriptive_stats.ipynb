## Analysis of Sepal

```python
! pip install seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sepal_length_setosa = iris_dataset.data[:, 0][0:50]
sepal_width_setosa = iris_dataset.data[:, 1][0:50]

sepal_length_versicolor = iris_dataset.data[:, 0][50:100]
sepal_width_versicolor = iris_dataset.data[:, 1][50:100]

sepal_length_virginica = iris_dataset.data[:, 0][100:150]
sepal_width_virginica = iris_dataset.data[:, 1][100:150]

sepal_length = iris_dataset.data[:, 0]
sepal_width = iris_dataset.data[:, 1]

# plot histograms

plt.figure(figsize=(10,5))

plt.subplot(121)
plt.title("Sepal Length")
sns.distplot(sepal_length, kde=False)

plt.subplot(122)
plt.title("Sepal Width")
sns.distplot(sepal_width, kde=False)
plt.savefig('plots.png')

print ("MEAN OF SEPAL LENGTH: "), np.mean(sepal_length)
print ("MEDIAN OF SEPAL LENGTH: "), np.median(sepal_length)
print ("=====================================")
print ("MEAN OF SEPAL WIDTH: "), np.mean(sepal_width)
print ("MEDIAN OF SEPAL WIDTH: "), np.median(sepal_width)
```

```python
# Relationship between Sepal Length and Width by species
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.scatter(x=sepal_length_setosa, y=sepal_width_setosa, c='#3DFC05', label='Setosa')
plt.scatter(x=sepal_length_versicolor, y=sepal_width_versicolor, c='#EE2421', label='Versicolor')
plt.scatter(x=sepal_length_virginica, y=sepal_width_virginica, c='#E8EE21', label='Virginica')
plt.legend()
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Relationship between Sepal Length and Width")

plt.savefig('scatter_plot.png')
```

## Analysis of Petal

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))


petal_length_setosa = iris_dataset.data[:, 2][0:50]
petal_width_setosa = iris_dataset.data[:, 3][0:50]

petal_length_versicolor = iris_dataset.data[:, 2][50:100]
petal_width_versicolor = iris_dataset.data[:, 3][50:100]

petal_length_virginica = iris_dataset.data[:, 2][100:150]
petal_width_virginica = iris_dataset.data[:, 3][100:150]

petal_length = iris_dataset.data[:, 2]
petal_width = iris_dataset.data[:, 3]

# plot histograms

plt.figure(figsize=(10,5))

plt.subplot(121)
plt.title("Petal Length")
sns.distplot(petal_length, bins=8, kde=False)

plt.subplot(122)
plt.title("Petal Width")
sns.distplot(petal_width, bins=7, kde=False)

plt.savefig('petal_plot.png')

print ("MEAN OF PETAL LENGTH: "), np.mean(petal_length)
print ("MEDIAN OF PETAL LENGTH: "), np.median(petal_length)
print ("=====================================")
print ("MEAN OF PETAL WIDTH: "), np.mean(petal_width)
print ("MEDIAN OF PETAL WIDTH: "), np.median(petal_width)
```

```python
# Relationship between Petal Length and Width by species
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))


plt.scatter(x=petal_length_setosa, y=petal_width_setosa, c='#3DFC05', label='Setosa')
plt.scatter(x=petal_length_versicolor, y=petal_width_versicolor, c='#EE2421', label='Versicolor')
plt.scatter(x=petal_length_virginica, y=petal_width_virginica, c='#E8EE21', label='Virginica')
plt.legend()
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Relationship between Petal Length and Width")

plt.savefig('scatter_plot.png')
```
